```python
import pandas as pd
import numpy as np

data_full=pd.read_csv('wine_dataset.csv')
data_full.head()
```

```python
data_full.describe()
```

```python
%matplotlib inline
data_full['quality'].hist()
```

```python
data_need=data_full[:]
del data_need['quality']
data_target=data_full['quality']
data_need.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data_need,data_target,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

# Train the model using the training sets
model.fit(X_train, y_train)

#Predict Output
predicted= model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predicted)*100)
```
